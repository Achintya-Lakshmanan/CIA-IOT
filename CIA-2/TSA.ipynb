```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("aq_milan_summer_2020.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data_new = data.copy()
```

```python
data["local_datetime"] = pd.to_datetime(data["local_datetime"])
data = data.set_index("local_datetime")
```

```python
data
```

```python
roll_mean = data.pm2p5.rolling(5).mean()
roll_std = data.pm2p5.rolling(5).std()

plt.figure(figsize = (15, 5))

plt.plot(data, label = "Conc")
plt.plot(roll_mean, label = "Moving avg")
plt.xlabel("Date")
plt.ylabel("ug/m3")

plt.legend()
plt.title("Conc of PM2.5")
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

dec_add = seasonal_decompose(data, model = "additive")

dec_add.plot()
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(data, lags = 50)
plot_pacf(data, lags = 50)
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

adf = adfuller(data)
print("ADF:", adf[0])

print("p-value:", adf[1])

print("\nCritical Values:\n")
for key, value in adf[4].items():
    print(key, value, sep = ": ")
```

From ADF test, data is stationary

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(data, order = (1, 0, 1), seasonal_order = (1, 0, 1, 24))
model_fit = model.fit()
```

```python
plt.figure(figsize = (15, 5))

plt.plot(data, c = 'red', label = 'Original')
plt.plot(model_fit.predict(40*24), c = 'blue', label = 'Predicted')
plt.legend()
plt.show()
```
